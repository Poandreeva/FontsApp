```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла training_log.csv
history_df = pd.read_csv('training_log.csv')

# Отрисовка графиков
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(history_df['loss'], label='Обучающая выборка', color='blue', linewidth=3)
plt.plot(history_df['val_loss'], label='Тестовая выборка', color='red', linewidth=3)
plt.title('Потери в зависимости от эпохи', fontsize=18)
plt.xlabel('Эпохи', fontsize=14)
plt.ylabel('Потери', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_df['accuracy'], label='Обучающая выборка', color='blue', linewidth=3)
plt.plot(history_df['val_accuracy'], label='Тестовая выборка', color='red', linewidth=3)
plt.title('Доля правильных ответов в зависимости от эпохи', fontsize=18)
plt.xlabel('Эпохи', fontsize=14)
plt.ylabel('Доля правильных ответов', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()

# Сохранение графиков
plt.savefig('training_plots.png')
plt.show()
```
