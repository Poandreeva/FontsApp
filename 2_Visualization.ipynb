```python
import pandas as pd
import matplotlib.pyplot as plt


history_df = pd.read_csv('training_log.csv')

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.plot()
plt.plot(history_df['loss'], label=' ', color='blue', linewidth=3)
plt.plot(history_df['val_loss'], label=' ', color='red', linewidth=3)
plt.title('    ', fontsize=18)
plt.xlabel('', fontsize=14)
plt.ylabel('', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_df['accuracy'], label=' ', color='blue', linewidth=3)
plt.plot(history_df['val_accuracy'], label=' ', color='red', linewidth=3)
plt.title('      ', fontsize=18)
plt.xlabel('', fontsize=14)
plt.ylabel('  ', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()

plt.savefig('training_plots.png')
plt.show()
```
